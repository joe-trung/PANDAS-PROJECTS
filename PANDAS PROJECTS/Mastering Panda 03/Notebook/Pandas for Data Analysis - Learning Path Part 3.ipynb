# 4. DATAFRAME CONCATENATION

```python
# Creating a dataframe from a dictionary
# Let's define a dataframe with a list of employees with IDs = 1, 2, 3, 4, 5 
# Check this out: https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
raw_data = {'Employee ID':['1','2','3','4','5'],
           'First Name':['Nancy','Alex','Shep','Max','Allen'],
           'Last Name':['Rob','Ali','George','Mitch','Steve']}
employee1_df = pd.DataFrame(raw_data, columns = ['Employee ID','First Name','Last Name'])
employee1_df
```

```python
# Let's define another dataframe for a separate list of clients (IDs = 6, 7, 8, 9, 10)
raw_data = {'Employee ID':['6','7','8','9','10'],
           'First Name':['Bill','Dena','Sara','Heather','Holly'],
           'Last Name':['Christian','Moe','Steve','Bob','Michelle']}
employee2_df = pd.DataFrame(raw_data, columns = ['Employee ID','First Name','Last Name'])
employee2_df
```

```python
# Let's concatenate both dataframes #1 and #2
# Note that we now have employee IDs from 1 to 10
# Note that by default ignore_index has been set to False meaning indexes from both dataframes are kept unchanged
employee_df = pd.concat([employee1_df,employee2_df])
employee_df
```

```python
# Let's concatenate both dataframes #1 and #2
# Note that by setting ignore_index = True, the index has been automatically set to numeric and now ranges from 1 to 9
employee_df = pd.concat([employee1_df,employee2_df], ignore_index = True)
employee_df
```

```python
len(employee_df)
```

```python
# You can also use the append method to perform similar task
# Note that order matters!
employee_df = employee2_df.append(employee1_df, ignore_index = True)
employee_df
```
